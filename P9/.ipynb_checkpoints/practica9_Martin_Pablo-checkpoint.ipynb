```python
import numpy as np
from sklearn.datasets import make_blobs
from sklearn.preprocessing import MinMaxScaler, OneHotEncoder
import matplotlib.pyplot as plt
import pandas as pd
from sklearn import SGDRegressor
```

```python
class neuron_simple_perceptron(object):
    def __init__(self, n_inputs=2, W=None, b=None):
        self.n_inputs = n_inputs
        self.W = W
        self.b = b
        if self.W is None:
            self.W = np.random.rand(1,self.n_inputs)-0.5
        if self.b is None:
            self.b = np.random.rand()-0.5
        return
    
    def predict(self, x):
        return np.sign(self.W @ x.T + self.b).astype(int)

    def fit(self, X, D):
        if np.sum(self.predict(X) == D) != D.shape[0]:
            continuar = True
            p = 0
            while continuar:
                v = p % X.shape[0]
                x = X[v].reshape(1,-1)
                y = self.predict(x)
                d = D[v]
                if y != d:
                    self.W  += d * x
                    self.b += d
                    if np.sum(self.predict(X) == D) == D.shape[0]:
                        continuar = False
                p += 1
        return
    
    def get_params(self):
        return self.W, self.b
```

```python
X,y = make_blobs(n_samples=1000, centers=2, n_features=2, random_state=0, cluster_std=0.5)

scaler = MinMaxScaler().fit(X)
X = scaler.transform(X)

perceptron = neuron_simple_perceptron(n_inputs=2)
perceptron.fit(X,2*y-1)


W,b = perceptron.get_params()
x_values = np.array([np.min(X[:,0]),np.max(X[:,0])])
y_values = (-1/W[0,1]) * (W[0,0] * x_values + b) 


# w1 x1 + w2 x2 + theta = 0
# x2 = -theta/w2 - w1 X1/w2

# X1 = [0,1]
# X2 = [-theta/w2, -theta/w2 - w1/w2]
```

```python

plt.scatter(X[np.argwhere(y==0),0], X[np.argwhere(y==0),1], c='r')
plt.scatter(X[np.argwhere(y==1),0], X[np.argwhere(y==1),1], c='b')
plt.plot(x_values,y_values)
plt.grid()
```

```python
## Ejercicio 2
from sklearn.datasets import fetch_california_housing
from sklearn.linear_model import SGDRegressor
from sklearn.model_selection import KFold
from sklearn.metrics import accuracy_score,r2_score
import pandas as pd
```

```python
housing = fetch_california_housing()
```

```python
scaler = MinMaxScaler()     # Aunque realice una regresión, hay que hacer un escalado 0,1 porque es una red neuronal.
```

```python
X_housing = scaler.fit_transform(housing.data)
y_housing = housing.target 
X_housing.shape, y_housing.shape
```

```python
reg = SGDRegressor(tol=1e-8,max_iter = 5000)
```

Preparamos el método de validación cruzada

```python
K = 10
kf = KFold(n_splits=K)
```

```python
r_reg = pd.DataFrame(columns=['R2','Accuracy'])
k = 0
for train_index, test_index in kf.split(X_housing):
    reg = SGDRegressor(tol=1e-6, max_iter=5000)
    reg.fit(X_housing[train_index,:], y_housing[train_index])
    y_predict = reg.predict(X_housing[test_index,:])
    error = np.mean(np.abs((y_predict - y_housing[test_index])/y_housing[test_index]) < 0.2)
    r_reg.loc[k,:] = [r2_score(y_housing[test_index], y_predict), error]
    k += 1
r_reg.index.name = "Fold"
r_reg
```
